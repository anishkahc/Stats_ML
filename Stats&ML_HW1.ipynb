```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, multinomial, expon, norm

# Binomial Distribution
n = 10
p = 0.5
binom_data = np.random.binomial(n, p, 1000) #sample data

# Plotting the histogram of samples
plt.hist(binom_data, bins=np.arange(n + 1) - 0.5,  color='green', edgecolor = 'black')
plt.xlabel('Number of Successes')
plt.ylabel('Count')
plt.title(f'Binomial Samples (n={n}, p={p})')
plt.show()

# Fitting Maximum Likelihood Parameters
mean_sample = np.mean(binom_data)
p_est = mean_sample / n
n_est = n

print(f'Mean of samples: {mean_sample:.2f}')
print(f'Estimated Parameters: n={n_est}, p={p_est:.2f}')
```

```python
# Multinomial Distribution
n = 10  # number of trials
p = [0.1, 0.2, 0.3, 0.4]  # probability of each category
multinom_data = np.random.multinomial(n, p, 1000) #sample data

# Plotting the histogram of samples
for i in range(len(p)):
    plt.hist(multinom_data[:, i], bins=np.arange(n + 1) - 0.5,  color='green', edgecolor = 'black')
    plt.xlabel('Number of Category ' + str(i + 1))
    plt.ylabel('Count')
    plt.title(f'Multinomial Samples (n={n}, p={p})')
    plt.show()

# Fitting Maximum Likelihood Parameters
mean_samples = np.mean(multinom_data, axis=0)
p_est = mean_samples / n
n_est = n

print(f'Mean of samples: {mean_samples}')
print(f'Estimated Parameters: n={n_est}, p={p_est}')
```

```python
# Exponential Distribution
beta = 2  # rate parameter
rv = expon(scale=1/beta)
expon_data = rv.rvs(1000)

# Plotting the histogram of samples
plt.hist(expon_data, bins=np.linspace(0, 10, 21), color='green', edgecolor = 'black')
plt.xlabel('x')
plt.ylabel('Count')
plt.title(f'Exponential Samples (beta={beta})')
plt.show()

# Fitting Maximum Likelihood Parameters
mean_sample = np.mean(expon_data)
beta_est = 1 / mean_sample

print(f'Mean of samples: {mean_sample:.2f}')
print(f'Estimated Parameters: beta={beta_est:.2f}')
```

```python
# Normal Distribution
mu = 5  # mean
sigma = 2  # standard deviation
rv = norm(loc=mu, scale=sigma)
norm_data = rv.rvs(1000)

# Plotting the histogram of samples
plt.hist(norm_data, bins=50, color='green', edgecolor = 'black')
plt.xlabel('x')
plt.ylabel('Count')
plt.title(f'Normal Samples (mu={mu}, sigma={sigma})')
plt.show()

# Fitting Maximum Likelihood Parameters
mu_est = np.mean(norm_data)
sigma_est = np.std(norm_data)

print(f'Mean of samples: {mu_est:.2f}')
print(f'Standard Deviation of samples: {sigma_est:.2f}')
print(f'Estimated Parameters: mu={mu_est:.2f}, sigma={sigma_est:.2f}')
```
